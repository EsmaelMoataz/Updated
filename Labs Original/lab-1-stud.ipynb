# Example 1

Print the prime numbers from 1 to 100.

```python
# Example 1
for i in range(101):
  prime=True
  j=2
  while prime and j<i:
    if (i % (j))==0:
      prime=False
    j+=1
  if prime:
    print(i)
```

# Task 1

Modify previous example and define function for testing if the number is a prime number.

```python
def isPrime(i):
  if i <=1 :
    return False

  prime=True
  j=2

  while j <= i//2:
    if i % j == 0:
      prime=False
      break
    j+=1
  return prime


def prime_number(n):
  for i in range(n):
    if isPrime(i):
      print(i)

prime_number(20)
```

# Task 2

Modify the previous task and define class finding the prime numbers in selected range.


```python
class Prime:
  def __init__(self, n):
    self.n = n
    self.primes = []

  def find_primes(self):
    for i in range(self.n):
      if self.isPrime(i):
        self.primes.append(i)

  def isPrime(self, num):
    prime=True
    j=2
    while prime and j<num:
      if (num % (j))==0:
        prime=False
      j+=1
    return prime


c = Prime(n=30)
c.find_primes()
print(c.primes)
```

# Task 9

The sigmoid function is frequently used activation of a neuron. It is defined by

$$y=\sigma(x)=\frac{1}{1+e^{-x}}$$

Draw a graf of this function for $x\in <-8,8>$.

```python
def sigmoid(x):
  x = np.array(x)
  return 1/(1+np.exp(-x))

# generate range of numbers
x = np.linspace(-8,8,100)

# plt.plot([1,2,3], sigmoid([3,5,6]))
# plt.show()

plt.axhline(y=0, color="gray", linestyle=":")
plt.axhline(y=0.5, color="gray", linestyle=":")
plt.axhline(y=1.0, color="gray", linestyle=":")
plt.axvline(color="gray")
plt.plot(x, sigmoid(x), label="$\sigma(x)= \frac{1}+e^{-x}}$")

plt.xlim(-8,8)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(fontsize=14)
plt.plot()
```

# Task 10

The RELU function is frequently used activation function of neurons in deep neural networks. It is defined by

$$y=\mbox{relu}(x)=\mbox{max}(0, x)$$

Draw a graf of this function for $x\in <-10,10>$.

```python
#np.meshgrid



def relu(x):
  x = np.array(x)
  x[x<0]=0
  return x

# generate range of numbers
x = np.linspace(-8,8,100)

# plt.plot([1,2,3], sigmoid([3,5,6]))
# plt.show()

plt.axhline(y=0, color="gray", linestyle=":")
plt.axvline(color="gray")
plt.plot(x, relu(x), label="$relu(x)= max(0,x)$")

plt.xlim(-8,8)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(fontsize=14)
plt.plot()
```

# Task 11

The Gaussian function is frequently used as an activation function of RBF neurons. It is defined by the equation

$$y=g(x)=e^{\frac{-x^2}{2\sigma^2}}$$

Draw a graf of this function for $x\in <-10,10>$ and $\sigma=\{0.2,0.5,1,2,5\}$.

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian(x, alpha):
  y = np.exp(-(x*x)/(2*alpha*alpha))
  return y

# generate range of numbers
x = np.linspace(-10,10,100)


plt.axhline(y=0, color="gray", linestyle=":")
for alpha in [0.2,0.5,1,2,5]:
  plt.plot(x, gaussian(x, alpha), label=f"g(x,{alpha})")

# plt.xlim(-10,10)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(fontsize=14)
plt.plot()
```

# Task 12

Draw response of the two input perceptron neuron defined by

$$y=\sigma(\sum_{i=0}^{1}w_ix_i + b)$$

where $w_0=10$,$w_1=10$, $b=-10$.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
w0, w1 = 10, 10
b = -10

# Sigmoid activation function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Create a grid of input values
x0 = np.linspace(0, 1, 100)
x1 = np.linspace(0, 1, 100)
X0, X1 = np.meshgrid(x0, x1)

# Compute the perceptron output
Z = sigmoid(w0 * X0 + w1 * X1 + b)

# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X0, X1, Z, cmap='viridis', edgecolor='none')

# Add labels and title


ax.set_xlabel('$x_0$')
ax.set_ylabel('$x_1$')
ax.set_zlabel('Output $y$')
plt.title('Response of the Two-Input Perceptron Neuron')
plt.show()
```
